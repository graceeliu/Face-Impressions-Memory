```python
import pandas as pd
import glob
import os
```

```python
#subs = [i for i in range(27,33)]
subs = [1]
runs = [str(i) for i in range(1, 7)]
```

```python
for i in subs:

    # changes 1 to '001'... how we have named the data!
    sub = str(i).zfill(3)

    # odd numbers are version 1, even numbers are version 2
    if i % 2 != 0:
        version = '1'
    else:
        version = '2'

    # get dfs for all runs
    for run in runs:
        # load df for sub, run
        temp = pd.read_csv(f'/Users/graceliu/Desktop/ColumbiaThesis/data/raw_data/{sub}/RUNv{version}_{sub}_run{run}.csv')

        # get onset and duration of events
        temp['onset'] = round(temp['global_time_start']).astype(int)
        temp['duration'] = temp['end_time'] - temp['onset']

        # get Black or White trial types... and fixation... and Probes
        # temp['trial_type'] = temp['Image'].dropna().apply(lambda x: 'BM' if 'BM' in x else ('WM' if 'WM' in x else 'Fix'))
        temp['trial_type'] = temp['Image'].dropna().apply(lambda x: 'BM' if 'BM' in x else 'WM')
        temp['trial_type'] = temp['trial_type'].fillna('Fix')
        temp.loc[temp['proc_type'] == 'ProbeProc', 'trial_type'] = 'Probe'
        temp = temp[temp['trial_type'] != 'Fix'].reset_index(drop=True)

        # select only desired cols
        temp = temp[['onset', 'trial_type', 'duration']]

        # # save to file
        save_dir = f'clean_data/sub-{sub}'
        if not os.path.exists(save_dir):
            os.makedirs(save_dir)   
        temp.to_csv(f'{save_dir}/sub-{sub}_catlocalizer_run{run}_events.csv', index=False)
```
